```python
import matplotlib.pyplot as plt

vehiculos=['Mazda cx 30','Renault kwi', 'Chevrolet Onix', 'Renault stepway', 'Renault duster', 'Renault sandero']
unidades=[4976,4713,3890,3705,3653,3469]
caballos=[186,66,160,111,143,156]
precios=[108,44,52.7,50,74,48.8]

colores=['red','orange','blue','green','white','yellow']
plt.pie(unidades, labels=vehiculos, colors=colores, shadow=True, startangle=90,autopct='%1.1f%%' ,wedgeprops={'edgecolor':'black'})
plt.title('grafica unidades vendidas de los vehiculos')
```
